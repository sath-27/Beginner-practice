Importing Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Importing dataset

```python
data = pd.read_csv('Social_Network_Ads.csv')
X = data.iloc[: , :-1].values
Y = data.iloc[: , -1].values
```

Splitting data into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state = 69) 
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(Y_train)
```

```python
print(Y_test)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

Training Logistic Regression Model on training set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=69)
classifier.fit(X_train , Y_train)
```

Predicting New Result

```python
print(classifier.predict(sc.transform([[30, 87000]])))
```

Predicting the test set results

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), Y_test.reshape(len(Y_test), 1)), 1)) 
```

Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test , y_pred)
print (cm)
print(accuracy_score(Y_test, y_pred))
```

Training KNN Model on training set

```python
from sklearn.neighbors import KNeighborsClassifier
classifier1 = KNeighborsClassifier(n_neighbors = 5 , metric = 'minkowski' , p=2)
classifier1.fit(X_train, Y_train)
```

Predicting New result

```python
print(classifier1.predict(sc.transform([[30, 87000]])))
```

Predicting the test set results

```python
y_pred1 = classifier1.predict(X_test)
print(np.concatenate((y_pred1.reshape(len(y_pred1), 1), Y_test.reshape(len(Y_test), 1)), 1)) 
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
cm = confusion_matrix(Y_test , y_pred1)
print(cm)
print(accuracy_score(Y_test , y_pred1))
```

Training SVM on the training data

```python
from sklearn.svm import SVC
classifier2 = SVC(kernel = 'linear' , random_state = 0)
classifier2.fit(X_train, Y_train)
```

Predicting new result

```python
print(classifier2.predict(sc.transform([[30, 87000]])))
```

Predicting the test set results

```python
y_pred2 = classifier2.predict(X_test)
print(np.concatenate((y_pred1.reshape(len(y_pred1), 1), Y_test.reshape(len(Y_test), 1)), 1))
```

Training Decision tree classifier on training data

```python
from sklearn.tree import DecisionTreeClassifier
classifier3 = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier3.fit(X_train, Y_train)
```

Predicting new result

```python
print(classifier3.predict(sc.transform([[30,87000]])))
```

Predicting test set results

```python
y_pred3 = classifier3.predict(X_test)
print(y_pred3)
```
